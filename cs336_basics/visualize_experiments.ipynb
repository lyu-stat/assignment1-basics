### Visualize validation loss vs learning rate

```python
from optimizer import learning_rate_schedule
import pandas as pd

lr_max = 3e-2
lr_min = 3e-3
w_step = 45
c_step = 1500
steps = list(range(1500))
lrs = [
    learning_rate_schedule(step, lr_max, lr_min, w_step, c_step) * 1e3 for step in steps
]
# Scale for better visualization

val_losses = pd.read_csv("./experiment_data/glamorous_feature_28_validation_loss.csv")[
    ["Step", "glamorous-feather-28 - validation_loss"]
]

fig, ax1 = plt.subplots()

# First scale (left y-axis): validation loss
ax1.plot(
    val_losses["Step"], val_losses["glamorous-feather-28 - validation_loss"], marker="o"
)
ax1.set_xlabel("step")
ax1.set_ylabel("validation loss")
ax1.grid(True)

# Second scale (right y-axis): learning rate
ax2 = ax1.twinx()  # share same x, different y
ax2.plot(steps, lrs, color="orange")
ax2.set_ylabel("learning rate")

plt.title("Validation loss and learning rate vs step")
plt.tight_layout()
plt.show()
```
